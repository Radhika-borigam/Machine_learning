```python
import pandas as pd
df=pd.read_csv("tvmarketing.csv");
print(df)
```

```python
print(df.head());
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(8,5))
plt.scatter(df['TV'], df['Sales'], color='blue', alpha=0.6)
plt.xlabel("TV Advertisement Budget ($1000)")
plt.ylabel("Sales (units in thousands)")
plt.title("Relationship Between TV Ad Budget and Sales")
plt.grid(False)
plt.show()
```

```python
X=df['TV']
Y=df['Sales']
n = len(X)
mean_X = sum(X) / n
mean_Y = sum(Y) / n
numerator = sum((X[i] - mean_X) * (Y[i] - mean_Y) for i in range(n))
denominator = sum((X[i] - mean_X) ** 2 for i in range(n))

b = numerator / denominator  
a = mean_Y - b * mean_X

print(f"Mean of X: {mean_X}")
print(f"Mean of Y: {mean_Y}")
print(f"Slope (b): {b}")
print(f"Intercept (a): {a}")


print(f"Regression Line: Y = {a:.2f} + {b:.2f}X")


```

```python
y_predict=a+b*20
print(y_predict)
```

```python
from sklearn.model_selection import train_test_split
X = df[['TV']]  
Y = df['Sales']  
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.3, random_state=42)
print("Shape of X_train:", X_train.shape)
print("Shape of y_train:", y_train.shape)
print("Shape of X_test:", X_test.shape)
print("Shape of y_test:", y_test.shape)
```

```python
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
lr.fit(X_train,y_train)
intercept=lr.intercept_
coefficients=lr.coef_
print(intercept,coefficients)
```

```python
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
lr.fit(X_train,y_train)
intercept=lr.intercept_
coefficients=lr.coef_
print(intercept,coefficients)
```

```python
Y_predict=lr.predict(X_test)
print(Y_predict)
```

```python
plt.scatter(X,Y)
plt.plot(X_train,lr.predict(X_train),color="red")
plt.xlabel("tvs")
plt.ylabel("sales");
plt.show()
```

```python
from sklearn.metrics import r2_score,mean_squared_error
r_square=r2_score(y_test,Y_predict)
mse=mean_squared_error(y_test,Y_predict)
print(mse)
print(r_square)
```
